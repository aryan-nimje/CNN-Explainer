```python
import customtkinter as ctk

window = ctk.CTk()
window.geometry("1400x900")

frame = ctk.CTkFrame(window)
frame.pack(fill="both", expand=True, padx=20, pady=20)

# create labels/images
for row in range(5):

    input_img = ctk.CTkLabel(frame, text=f"Input {row}")
    input_img.grid(row=row, column=0, padx=20, pady=20)

    conv_img = ctk.CTkLabel(frame, text=f"Conv {row}")
    conv_img.grid(row=row, column=1, padx=20, pady=20)

    pool_img = ctk.CTkLabel(frame, text=f"Pool {row}")
    pool_img.grid(row=row, column=2, padx=20, pady=20)

    dense_img = ctk.CTkLabel(frame, text=f"Dense {row}")
    dense_img.grid(row=row, column=3, padx=20, pady=20)

window.mainloop()
```
