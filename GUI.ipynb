```python
import customtkinter as ctk
import tkinter as tk
from PIL import Image, ImageDraw

#==========================================================================================================================================#

#Frame Architecture Start
window = ctk.CTk()
window.geometry("1400x900")

scroll_frame = ctk.CTkScrollableFrame(window)
scroll_frame.pack(fill="both", expand=True)

content_frame = ctk.CTkFrame(
    scroll_frame,
    width=2000,
    height=3000
)
content_frame.pack()
#Frame Architecture End

#==========================================================================================================================================#

## Title Start Markdowns
label = ctk.CTkLabel(
    content_frame,
    text = "CNN Explainer",
    font = ("Arial", 32, "bold")
)
label.pack(pady = 20)

task = ctk.CTkLabel(
    content_frame,
    text = "Draw an Image of either of these objects: ",
    font = ("Arial", 20, "italic")
)
task.pack(pady = 10)

obj = ctk.CTkLabel(
    content_frame,
    text = "Shoes, T-Shirt, Dress, Jeans, Bags"
)
obj.pack(pady = 5)
## Title Markdowns End

#==========================================================================================================================================#

## Canvas Declaration Start
canvas = tk.Canvas(content_frame, bg = 'white', width = 500, height = 500)
canvas.pack(pady = 20)
## Canvas Declaration End

#==========================================================================================================================================#

## Image Declaration for canvas and saving Start
img = Image.new("RGB",  (500, 500),  "white")
draw = ImageDraw.Draw(img)

last_x, last_y = None, None

def paint(event):
    global last_x, last_y

    if last_x is not None and last_y is not None:
        canvas.create_line(
            last_x, last_y,
            event.x,  event.y,
            width = 4,
            capstyle = tk.ROUND,
            smooth = True
        )
        draw.line(
            [last_x, last_y, event.x, event.y],
            width = 4,
            fill = "black"
        )

    last_x, last_y = event.x, event.y

def reset(event):
    global last_x, last_y
    last_x = None
    last_y = None

def save():
    img.save("Drawing.png")
    print("Image Saved!!!")
## Image Declaration for canvas and saving End

#==========================================================================================================================================#


canvas.bind("<B1-Motion>", paint)
canvas.bind("<ButtonRelease-1>", reset)

save_btn = ctk.CTkButton(content_frame, text = "SaveImage", command = save)
save_btn.pack(pady = 10)



content = ctk.CTkFrame(content_frame, width=2000, height=3000)
content.pack()

#==========================================================================================================================================#

## Input Image Channels Start
ip_1 = Image.open("Drawing.png")
ctk_ip_1 = ctk.CTkImage(
    light_image = ip_1,
    dark_image = ip_1,
    size = (75, 75) 
)
label_ip_1 = ctk.CTkLabel(
    content_frame,
    image = ctk_ip_1,
    text = ""
)
label_ip_1.image = ctk_ip_1
label_ip_1.place(x = 50, y = 950)

ip_2 = Image.open("Drawing.png")
ctk_ip_2 = ctk.CTkImage(
    light_image = ip_2,
    dark_image = ip_2,
    size = (75, 75) 
)
label_ip_2 = ctk.CTkLabel(
    content_frame,
    image = ctk_ip_2,
    text = ""
)
label_ip_2.image = ctk_ip_2
label_ip_2.place(x = 50, y =1150)

ip_3 = Image.open("Drawing.png")
ctk_ip_3 = ctk.CTkImage(
    light_image = ip_3,
    dark_image = ip_3,
    size = (75, 75) 
)
label_ip_3 = ctk.CTkLabel(
    content_frame,
    image = ctk_ip_3,
    text = ""
)
label_ip_3.image = ctk_ip_3
label_ip_3.place(x = 50, y =1350)
# Input Image Channels Over

#==========================================================================================================================================#

## Convolution 1 Channels Start
conv_1 = Image.open("Drawing.png")
ctk_conv_1 = ctk.CTkImage(
    light_image = conv_1,
    dark_image = conv_1,
    size = (75, 75) 
)
label_conv_1 = ctk.CTkLabel(
    content_frame,
    image = ctk_conv_1,
    text = ""
)
label_conv_1.image = ctk_conv_1
label_conv_1.place(x = 175, y = 700)
## Convolution 1 Channel End

#==========================================================================================================================================#

## ReLU 1 Channels Start
relu_1 = Image.open("Drawing.png")
ctk_relu_1 = ctk.CTkImage(
    light_image = relu_1,
    dark_image = relu_1,
    size = (75, 75) 
)
label_relu_1 = ctk.CTkLabel(
    content_frame,
    image = ctk_relu_1,
    text = ""
)
label_relu_1.image = ctk_relu_1
label_relu_1.place(x = 275, y = 700)
## ReLU 1 Channels End

#==========================================================================================================================================#

## Convolution 2 Channels Start
conv_2 = Image.open("Drawing.png")
ctk_conv_2 = ctk.CTkImage(
    light_image = conv_2,
    dark_image = conv_2,
    size = (75, 75) 
)
label_conv_2 = ctk.CTkLabel(
    content_frame,
    image = ctk_conv_2,
    text = ""
)
label_conv_2.image = ctk_conv_2
label_conv_2.place(x = 400, y = 700)
## Convolution 2 Channels End

#==========================================================================================================================================#

## ReLU 1 Channels Start
relu_2 = Image.open("Drawing.png")
ctk_relu_2 = ctk.CTkImage(
    light_image = relu_2,
    dark_image = relu_2,
    size = (75, 75) 
)
label_relu_2 = ctk.CTkLabel(
    content_frame,
    image = ctk_relu_2,
    text = ""
)
label_relu_2.image = ctk_relu_2
label_relu_2.place(x = 500, y = 700)
## ReLU 2 Channels End

#==========================================================================================================================================#

## MaxPool 1 Channels Start
mpool_1 = Image.open("Drawing.png")
ctk_mpool_1 = ctk.CTkImage(
    light_image = mpool_1,
    dark_image = mpool_1,
    size = (75, 75) 
)
label_mpool_1 = ctk.CTkLabel(
    content_frame,
    image = ctk_mpool_1,
    text = ""
)
label_mpool_1.image = ctk_mpool_1
label_mpool_1.place(x = 600, y = 700)
## MaxPool 2 Channels End

#==========================================================================================================================================#

## Convolution 3 Channels Start
conv_3 = Image.open("Drawing.png")
ctk_conv_3 = ctk.CTkImage(
    light_image = conv_3,
    dark_image = conv_3,
    size = (75, 75) 
)
label_conv_3 = ctk.CTkLabel(
    content_frame,
    image = ctk_conv_3,
    text = ""
)
label_conv_3.image = ctk_conv_3
label_conv_3.place(x = 725, y = 700)
## Convolution 1 Channel End

#==========================================================================================================================================#

## ReLU 1 Channels Start
relu_3 = Image.open("Drawing.png")
ctk_relu_3 = ctk.CTkImage(
    light_image = relu_3,
    dark_image = relu_3,
    size = (75, 75) 
)
label_relu_3 = ctk.CTkLabel(
    content_frame,
    image = ctk_relu_3,
    text = ""
)
label_relu_3.image = ctk_relu_3
label_relu_3.place(x = 825, y = 700)
## ReLU 1 Channels End

#==========================================================================================================================================#

## Convolution 2 Channels Start
conv_4 = Image.open("Drawing.png")
ctk_conv_4 = ctk.CTkImage(
    light_image = conv_4,
    dark_image = conv_4,
    size = (75, 75) 
)
label_conv_4 = ctk.CTkLabel(
    content_frame,
    image = ctk_conv_4,
    text = ""
)
label_conv_4.image = ctk_conv_4
label_conv_4.place(x = 950, y = 700)
## Convolution 2 Channels End

#==========================================================================================================================================#

## ReLU 1 Channels Start
relu_4 = Image.open("Drawing.png")
ctk_relu_4 = ctk.CTkImage(
    light_image = relu_4,
    dark_image = relu_4,
    size = (75, 75) 
)
label_relu_4 = ctk.CTkLabel(
    content_frame,
    image = ctk_relu_4,
    text = ""
)
label_relu_4.image = ctk_relu_4
label_relu_4.place(x = 1050, y = 700)
## ReLU 2 Channels End

#==========================================================================================================================================#

## MaxPool 1 Channels Start
mpool_2 = Image.open("Drawing.png")
ctk_mpool_2 = ctk.CTkImage(
    light_image = mpool_2,
    dark_image = mpool_2,
    size = (75, 75) 
)
label_mpool_2 = ctk.CTkLabel(
    content_frame,
    image = ctk_mpool_2,
    text = ""
)
label_mpool_2.image = ctk_mpool_2
label_mpool_2.place(x = 1150, y = 700)
## MaxPool 2 Channels End
window.mainloop()
```
