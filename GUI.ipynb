```python
import customtkinter as ctk
import tkinter as tk
from PIL import Image, ImageDraw

window = ctk.CTk()
window.geometry("1200x800")

scroll_frame = ctk.CTkScrollableFrame(window)
scroll_frame.pack(fill="both", expand=True)

label = ctk.CTkLabel(
    scroll_frame,
    text = "CNN Explainer",
    font = ("Arial", 32, "bold")
)
label.pack(pady = 20)

task = ctk.CTkLabel(
    scroll_frame,
    text = "Draw an Image of either of these objects: ",
    font = ("Arial", 20, "italic")
)
task.pack(pady = 10)

obj = ctk.CTkLabel(
    scroll_frame,
    text = "Shoes, T-Shirt, Dress, Jeans, Bags"
)
obj.pack(pady = 5)

canvas = tk.Canvas(scroll_frame, bg = 'white', width = 500, height = 500)
canvas.pack(pady = 20)

img = Image.new("RGB",  (500, 500),  "white")
draw = ImageDraw.Draw(img)

last_x, last_y = None, None

def paint(event):
    global last_x, last_y

    if last_x is not None and last_y is not None:
        canvas.create_line(
            last_x, last_y,
            event.x,  event.y,
            width = 4,
            capstyle = tk.ROUND,
            smooth = True
        )
        draw.line(
            [last_x, last_y, event.x, event.y],
            width = 4,
            fill = "black"
        )

    last_x, last_y = event.x, event.y

def reset(event):
    global last_x, last_y
    last_x = None
    last_y = None

def save():
    img.save("Drawing.png")
    print("Image Saved!!!")

canvas.bind("<B1-Motion>", paint)
canvas.bind("<ButtonRelease-1>", reset)

save_btn = ctk.CTkButton(scroll_frame, text = "SaveImage", command = save)
save_btn.pack(pady = 10)

img_1 = Image.open("Drawing.png")
ctk_img_1 = ctk.CTkImage(
    light_image = img_1,
    dark_image = img_1,
    size = (100, 100) 
)
label_img_1 = ctk.CTkLabel(
    scroll_frame,
    image = ctk_img_1,
    text = ""
)
label_img_1.place(y = 750,
                x = 10)

img_2 = Image.open("Drawing.png")
ctk_img_2 = ctk.CTkImage(
    light_image = img_2,
    dark_image = img_2,
    size = (100, 100) 
)
label_img_2 = ctk.CTkLabel(
    scroll_frame,
    image = ctk_img_2,
    text = ""
)
label_img_2.place(y = 900,
                  x = 10)

window.mainloop()
```
