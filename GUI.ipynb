# Model Class

```python
from torch.nn.modules.pooling import MaxPool2d
class ConvModel(nn.Module):
  def __init__(self, input_shape: int, hidden_units: int, output_shape: int):
    super().__init__()
    #Block 1 Start
    self.conv_1 = nn.Conv2d(in_channels = input_shape,
                           out_channels = hidden_units,
                           kernel_size = (2,2),
                           stride = 1,
                           padding = 1)
    self.relu_1 = nn.ReLU()

    self.conv_2 = nn.Conv2d(in_channels = hidden_units,
                           out_channels = hidden_units,
                           kernel_size = 1,
                           stride = 1,
                           padding = 1)
    self.relu_2 = nn.ReLU()
    self.maxpool_1 = nn.MaxPool2d(kernel_size = (2,2))
    #Block 1 End

    #Block 2 Start
    self.conv_3 = nn.Conv2d(in_channels = hidden_units,
                           out_channels = hidden_units,
                           kernel_size = (2,2),
                           stride = 1,
                           padding = 1)
    self.relu_3 = nn.ReLU()

    self.conv_4 = nn.Conv2d(in_channels = hidden_units,
                           out_channels = hidden_units,
                           kernel_size = 1,
                           stride = 1,
                           padding = 1)
    self.relu_4 = nn.ReLU()
    self.maxpool_2 = nn.MaxPool2d(kernel_size = (2,2))
    #Block 2 End

    self.classifier = nn.Sequential(
        nn.Flatten(),
        nn.Linear(in_features = hidden_units * 9* 9,
                  out_features = output_shape)
    )

  def forward(self, X):
    X = self.conv_1(X)
    X = self.relu_1(X)
    X = self.conv_2(X)
    X = self.relu_2(X)
    X = self.maxpool_1(X)

    X = self.conv_3(X)
    X = self.relu_3(X)
    X = self.conv_4(X)
    X = self.relu_4(X)
    X = self.maxpool_2(X)

    X = self.classifier(X)
    return X

model = ConvModel(input_shape = 1, 
                  hidden_units = 10,
                  output_shape = 10)
```

```python
import customtkinter as ctk
import tkinter as tk
from PIL import Image, ImageDraw

pallete = ctk.CTk()
pallete.geometry("800x700")


## Title Start Markdowns
label = ctk.CTkLabel(
    pallete,
    text = "Drawing Simulation",
    font = ("Arial", 32, "bold")
)
label.pack(pady = 20)

task = ctk.CTkLabel(
    pallete,
    text = "Draw an Image of either of these objects, and our CNN will predict what you drew!!!",
    font = ("Arial", 20, "italic")
)
task.pack(pady = 10)

obj = ctk.CTkLabel(
    pallete,
    text = "Shoes, T-Shirt, Dress, Jeans, Bags"
)
obj.pack(pady = 5)
## Title Markdowns End

canvas = tk.Canvas(pallete, bg = 'white', width = 500, height = 500)
canvas.pack(pady = 20)

img = Image.new("L",  (500, 500),  "white")
draw = ImageDraw.Draw(img)

last_x, last_y = None, None

def paint(event):
    global last_x, last_y

    if last_x is not None and last_y is not None:
        canvas.create_line(
            last_x, last_y,
            event.x, event.y,
            width = 3,
            capstyle = tk.ROUND,
            smooth = True
        )
        draw.line(
            [last_x, last_y, event.x, event.y],
            width = 3,
            fill = 'black'
        )

    last_x, last_y = event.x, event.y

def save():
    img.save("Testing.png")
    print("Pallete Drawing saved!!!")

def reset(event):
    global last_x, last_y

    last_x, last_y = None, None

canvas.bind("<B1-Motion>", paint)
canvas.bind("<ButtonRelease-1>", reset)

save_btn = ctk.CTkButton(pallete, text = "Save Image", command = save)
save_btn.pack(pady = 10)

pallete.mainloop()
```
