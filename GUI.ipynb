# Model Class

```python
from torch.nn.modules.pooling import MaxPool2d
class ConvModel(nn.Module):
  def __init__(self, input_shape: int, hidden_units: int, output_shape: int):
    super().__init__()
    #Block 1 Start
    self.conv_1 = nn.Conv2d(in_channels = input_shape,
                           out_channels = hidden_units,
                           kernel_size = (2,2),
                           stride = 1,
                           padding = 1)
    self.relu_1 = nn.ReLU()

    self.conv_2 = nn.Conv2d(in_channels = hidden_units,
                           out_channels = hidden_units,
                           kernel_size = 1,
                           stride = 1,
                           padding = 1)
    self.relu_2 = nn.ReLU()
    self.maxpool_1 = nn.MaxPool2d(kernel_size = (2,2))
    #Block 1 End

    #Block 2 Start
    self.conv_3 = nn.Conv2d(in_channels = hidden_units,
                           out_channels = hidden_units,
                           kernel_size = (2,2),
                           stride = 1,
                           padding = 1)
    self.relu_3 = nn.ReLU()

    self.conv_4 = nn.Conv2d(in_channels = hidden_units,
                           out_channels = hidden_units,
                           kernel_size = 1,
                           stride = 1,
                           padding = 1)
    self.relu_4 = nn.ReLU()
    self.maxpool_2 = nn.MaxPool2d(kernel_size = (2,2))
    #Block 2 End

    self.classifier = nn.Sequential(
        nn.Flatten(),
        nn.Linear(in_features = hidden_units * 9* 9,
                  out_features = output_shape)
    )

  def forward(self, X):
    X = self.conv_1(X)
    X = self.relu_1(X)
    X = self.conv_2(X)
    X = self.relu_2(X)
    X = self.maxpool_1(X)

    X = self.conv_3(X)
    X = self.relu_3(X)
    X = self.conv_4(X)
    X = self.relu_4(X)
    X = self.maxpool_2(X)

    X = self.classifier(X)
    return X

model = ConvModel(input_shape = 1, 
                  hidden_units = 10,
                  output_shape = 10)
```

# Canvas

```python
import customtkinter as ctk
import tkinter as tk
from PIL import Image, ImageDraw

pallete = ctk.CTk()
pallete.geometry("900x750")

# ---------------- TITLE ---------------- #

label = ctk.CTkLabel(
    pallete,
    text="Drawing Simulation",
    font=("Arial", 32, "bold")
)
label.pack(pady=20)

task = ctk.CTkLabel(
    pallete,
    text="Draw an Image of either of these objects, and our CNN will predict what you drew!!!",
    font=("Arial", 20, "italic")
)
task.pack(pady=10)

obj = ctk.CTkLabel(
    pallete,
    text="Shoes, T-Shirt, Dress, Jeans, Bags"
)
obj.pack(pady=5)

# ---------------- CURRENT COLOR ---------------- #

current_color = "black"

# ---------------- MAIN FRAME ---------------- #

main_frame = ctk.CTkFrame(pallete)
main_frame.pack(pady=20)

# ---------------- COLOR BUTTONS ---------------- #

color_frame = ctk.CTkFrame(main_frame)
color_frame.grid(row=0, column=0, padx=20)

def change_color(new_color):
    global current_color
    current_color = new_color

colors = ["white", "yellow", "red", "pink", "green", "blue", "black"]

for color in colors:
    btn = ctk.CTkButton(
        color_frame,
        text="",
        width=40,
        height=40,
        fg_color=color,
        hover_color=color,
        command=lambda c=color: change_color(c)
    )
    btn.pack(pady=10)

# ---------------- CANVAS ---------------- #

canvas = tk.Canvas(
    main_frame,
    bg='white',
    width=500,
    height=500
)
canvas.grid(row=0, column=1)

# PIL Image for saving
img = Image.new("RGB", (500, 500), "white")
draw = ImageDraw.Draw(img)

last_x, last_y = None, None

# ---------------- DRAW FUNCTION ---------------- #

def paint(event):
    global last_x, last_y

    if last_x is not None and last_y is not None:

        canvas.create_line(
            last_x,
            last_y,
            event.x,
            event.y,
            width=5,
            fill=current_color,
            capstyle=tk.ROUND,
            smooth=True
        )

        draw.line(
            [last_x, last_y, event.x, event.y],
            fill=current_color,
            width=5
        )

    last_x, last_y = event.x, event.y

# ---------------- RESET ---------------- #

def reset(event):
    global last_x, last_y
    last_x, last_y = None, None

# ---------------- SAVE ---------------- #

def save():
    img.save("Testing.png")
    print("Palette Drawing Saved!!!")

# ---------------- CLEAR CANVAS ---------------- #

def clear_canvas():
    canvas.delete("all")

    global img, draw
    img = Image.new("RGB", (500, 500), "white")
    draw = ImageDraw.Draw(img)

# ---------------- BINDINGS ---------------- #

canvas.bind("<B1-Motion>", paint)
canvas.bind("<ButtonRelease-1>", reset)

# ---------------- BUTTONS ---------------- #

save_btn = ctk.CTkButton(
    pallete,
    text="Save Image",
    command=save
)
save_btn.pack(pady=10)

clear_btn = ctk.CTkButton(
    pallete,
    text="Clear Canvas",
    command=clear_canvas
)
clear_btn.pack(pady=10)

# ---------------- MAINLOOP ---------------- #

pallete.mainloop()
```
